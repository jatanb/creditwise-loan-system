```python
df=pd.read_csv("loan.csv")


cc=df.select_dtypes(include=["object"]).columns
nc=df.select_dtypes(include=["number"]).columns

num_i=SimpleImputer(strategy="mean")
df[nc]=num_i.fit_transform(df[nc])
cat_i=SimpleImputer(strategy="most_frequent")
df[cc]=cat_i.fit_transform(df[cc])

le=LabelEncoder()
df["Loan_Approved"]=le.fit_transform(df["Loan_Approved"])
df["Education_Level"]=le.fit_transform(df["Education_Level"])

cols=["Employment_Status","Marital_Status","Loan_Purpose","Property_Area","Gender","Employer_Category"]
ohe=OneHotEncoder(drop="first",sparse_output=False,handle_unknown="ignore")
encoded=ohe.fit_transform(df[cols])
encod_df=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(cols),index=df.index)
df=pd.concat([df.drop(columns=cols),encod_df],axis=1)
df.head()

```

```python
x=df.drop(columns={"Loan_Approved","Applicant_ID"})
y=df["Loan_Approved"]

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

scaler=StandardScaler()
x_train_scale=scaler.fit_transform(x_train)
x_test_scale=scaler.transform(x_test)


model=GaussianNB()
model.fit(x_train_scale,y_train)
```

```python
y_pred=model.predict(x_test_scale)
y_pred
```

```python
print("precision :",precision_score(y_test,y_pred))
print("f1_score :",f1_score(y_test,y_pred))
print("recall :",recall_score(y_test,y_pred))
print("accuracy :",accuracy_score(y_test,y_pred))
print("cm :",confusion_matrix(y_test,y_pred))
```

```python
y_pred=model.predict(x_test_scale)


print("precision :",precision_score(y_test,y_pred))
print("f1_score :",f1_score(y_test,y_pred))
print("recall :",recall_score(y_test,y_pred))
print("accuracy :",accuracy_score(y_test,y_pred))
print("cm :",confusion_matrix(y_test,y_pred))
```

```python
df["squred_credit"]=df["Credit_Score"]**2
df["squred_dti"]=df["DTI_Ratio"]**2

x=df.drop(columns={"Loan_Approved","Applicant_ID","Credit_Score","DTI_Ratio"})
y=df["Loan_Approved"]


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

scaler=StandardScaler()
x_train_scale=scaler.fit_transform(x_train)
x_test_scale=scaler.transform(x_test)


model=GaussianNB()
model.fit(x_train_scale,y_train)
```

```python
y_pred=model.predict(x_test_scale)


print("precision :",precision_score(y_test,y_pred))
print("f1_score :",f1_score(y_test,y_pred))
print("recall :",recall_score(y_test,y_pred))
print("accuracy :",accuracy_score(y_test,y_pred))
print("cm :",confusion_matrix(y_test,y_pred))
```

```python
model=DecisionTreeClassifier(max_depth=4)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print("accuracy :",accuracy_score(y_pred,y_test))
print("precision :",precision_score(y_pred,y_test))
```

```python
rf=RandomForestClassifier(n_estimators=201,oob_score=True)
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
print(accuracy_score(y_pred,y_test))
```

```python
base_model=DecisionTreeClassifier()
bagg=BaggingClassifier(
    base_model,
    n_estimators=201
)
bagg.fit(x_train,y_train)
```

```python
y_pred=bagg.predict(x_test)
print(accuracy_score(y_pred,y_test))
```

```python
grb=GradientBoostingClassifier(
    max_depth=10,
    n_estimators=401,
    learning_rate=0.1,
    random_state=42
)
grb.fit(x_train,y_train)
```

```python
y_pred=grb.predict(x_test)
print(accuracy_score(y_pred,y_test))
```

```python
svm=SVC()
lr=LogisticRegression()
drb=DecisionTreeClassifier(max_depth=9)

vot=VotingClassifier(
  estimators=[
   ("svm",svm),
   ("lr",lr),
   ("drb",drb),
  ]
)
vot.fit(x_train,y_train)
```

```python
y_pred=vot.predict(x_test)
print("precision :",precision_score(y_test,y_pred))
print("f1_score :",f1_score(y_test,y_pred))
print("recall :",recall_score(y_test,y_pred))
print("accuracy :",accuracy_score(y_test,y_pred))
print("cm :",confusion_matrix(y_test,y_pred))
```
